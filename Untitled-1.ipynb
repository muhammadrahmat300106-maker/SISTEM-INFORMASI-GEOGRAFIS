```python
import pandas as pd
import geopandas as gpd
import folium
```

```python
data = {
    "nama": ["ABC", "DEF"],
    "latitude": [-5.216249, -5.216249],
    "longitude": [112.268818, 120.523803]
}

df = pd.DataFrame(data)
df
```

```python
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df.longitude, df.latitude),
    crs="EPSG:4326"
)
```

```python
lat = -5.162754
long = 122.858770
```

```python
m = folium.Map(location=[lat, long], zoom_start=12)
```

```python
folium.Marker(
    location=[-5.250702, 122.925010],
    tooltip="pasarwajo"
).add_to(m)

folium.Marker(
    location=[-5.320634, 123.030955],
    tooltip="buton"
).add_to(m)

folium.Marker(
    location=[-5.470115, 122.629129],
    tooltip="bau-bau"
).add_to(m)

folium.Marker(
    location=[-5.623204, 122.623078],
    tooltip="batauga"
).add_to(m)

folium.Marker(
    location=[-5.075299, 122.509980],
    tooltip="muna"
).add_to(m)

```

```python

koordinat_jalan = [
    [-5.075299, 122.509980],  # muna
    [-5.470115, 122.629129],  # bau-bau
    [-5.623204, 122.623078],  # batauga
    [-5.250702, 122.925010],  # pasarwajo
    [-5.320634, 123.030955]   # buton
]


folium.PolyLine(
    locations=koordinat_jalan,
    color="green",            
    weight=4,                
    opacity=0.8,             
    tooltip="Rute Perjalanan"
).add_to(m)
```

```python
koordinat_wilayah = [
    [-5.075299, 122.509980],  
    [-5.470115, 122.629129],  
    [-5.623204, 122.623078],  
    [-5.320634, 123.030955],  
    [-5.250702, 122.925010]   
]

folium.Polygon(
    locations=koordinat_wilayah,
    color="red",              
    weight=2,                 
    fill=True,                
    fill_color="red",         
    fill_opacity=0.3,         
    tooltip="Wilayah Cakupan"
).add_to(m)
```

```python
m.save('mapl.html')
```
